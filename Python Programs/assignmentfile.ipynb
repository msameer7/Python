## TASK 1.1 Remove Duplications

```python
def remove_duplicates_file(input_path, output_path):
    # Read in the numbers from the input file
    with open(input_path, 'r') as input_file:
        lines = input_file.readlines()
        # Split the lines into separate numbers
        numbers = [int(num) for line in lines for num in line.strip().split()]

    # Sort the input list
    numbers = sorted(numbers)

    # Initialize variables
    current_num = None
    count = 0
    last_index = -1
    output = []

    # Iterate over the sorted list
    for i, num in enumerate(numbers):
        # If the number is different from the current number, add the current number to the output list
        # based on its count and last index. Then, reset the current number, its count, and its last index
        # to the values of the new number.
        if num != current_num:
            if count == 1:
                output.append(current_num)
            elif count % 2 == 1:
                output.append(numbers[last_index])
            else:
                output.append(numbers[last_index + 1])

            current_num = num
            count = 1
            last_index = i
        # If the number is the same as the current number, increment its count and update its last index.
        else:
            count += 1

    # Add the final number to the output list based on its count and last index.
    if count == 1:
        output.append(current_num)
    elif count % 2 == 1:
        output.append(numbers[last_index])
    else:
        output.append(numbers[last_index + 1])

    # Write the output list to the output file
    with open(output_path, 'w') as output_file:
        output_file.write('\n'.join(str(num) for num in output))
    
    # Print the output list
    print(output)
```

## TASK 1.2 Operations and Numbers

```python
import time

def read_and_sort_integers(file_path):
    with open(file_path, 'r') as file:
        integers = [int(num) for line in file for num in line.split()]
    integers.sort()
    return integers
```

```python
# Define a function to search for a value in a list
def search_value(value, integers):
    return value in integers
```

```python
# Define a function to insert a value into a list in ascending order
def insert_value(value, integers):
    index = 0
    while index < len(integers) and integers[index] < value:
        index += 1
    integers.insert(index, value)
```

```python
# Define a function to delete a value from a list (including duplicates)
def delete_value(value, integers):
    while value in integers:
        integers.remove(value)
```

```python
# Define a function to perform data operations and write the results to an output file
def perform_data_operations(operations, integers, output_file):
    with open(output_file, 'w') as file:
        for operation in operations:
            if operation[0] == 1:  # Search
                start_time = time.time()
                result = search_value(operation[1], integers)
                end_time = time.time()
                output_str = f"Search for {operation[1]}: {result}. Time taken: {end_time - start_time} seconds.\n"
                file.write(output_str)
            elif operation[0] == 2:  # Insertion
                start_time = time.time()
                insert_value(operation[1], integers)
                end_time = time.time()
                output_str = f"Insertion of {operation[1]} completed. Time taken: {end_time - start_time} seconds.\n"
                file.write(output_str)
            elif operation[0] == 3:  # Deletion
                start_time = time.time()
                delete_value(operation[1], integers)
                end_time = time.time()
                output_str = f"Deletion of {operation[1]} completed. Time taken: {end_time - start_time} seconds.\n"
                file.write(output_str)
```

```python
if __name__ == '__main__':
    integers = read_and_sort_integers('task1_2_operations.txt')
    operations = [(1, 10), (2, 20), (3, 30)]  # Example operations
    perform_data_operations(operations, integers, 'output1.2(2).txt')
    print("Final state of integers list:", integers)
```

## TASK 1.3 (search with parallelism)

```python
import threading

# Function to count the frequency of each name in a chunk of words
def count_names(chunk, names, results, results_lock):
    frequency = {}
    for name in names:
        frequency[name] = 0
    for word in chunk:
        if word in frequency:
            frequency[word] += 1
    # Acquire a lock before modifying the results dictionary
    with results_lock:
        for name in names:
            results[name] += frequency[name]
```

```python
if __name__ == '__main__':
    # Read the text file and split it into a list of words
    with open('task1_3_text.txt', 'r', encoding='utf-8') as f:
        words = f.read().split()

    # Read the names file and create a list of names
    with open('task1_3_names.txt', 'r') as f:
        names = f.read().splitlines()

    # Split the list of words into multiple chunks
    num_threads = 4
    chunk_size = len(words) // num_threads
    word_chunks = [words[i:i+chunk_size] for i in range(0, len(words), chunk_size)]

    # Create a results dictionary and a lock for synchronizing access to it
    results = {}
    for name in names:
        results[name] = 0
    results_lock = threading.Lock()

    # Create and start a thread for each chunk of words
    threads = []
    for chunk in word_chunks:
        thread = threading.Thread(target=count_names, args=(chunk, names, results, results_lock))
        thread.start()
        threads.append(thread)

    # Wait for all the threads to complete
    for thread in threads:
        thread.join()

    # Write the frequency dictionary to a new file
    with open('frequency_file.txt', 'w') as f:
        for name in names:
            f.write('{} {}\n'.format(name, results[name]))
```

## TASK 1.4 (Railway Cheapest Cost)

```python
import csv
from collections import defaultdict
import heapq

def load_railway_network():
    railway_network = defaultdict(dict)
    with open('task1_4_railway_network.csv', 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            start, end, cost = row
            cost = int(cost)
            railway_network[start][end] = cost
            railway_network[end][start] = cost
    return railway_network
```

```python
def dijkstra(railway_network, start, end):
    heap = [(0, start)]
    visited = set()
    while heap:
        (cost, current) = heapq.heappop(heap)
        if current in visited:
            continue
        visited.add(current)
        if current == end:
            return cost
        for neighbor, neighbor_cost in railway_network[current].items():
            if neighbor not in visited:
                heapq.heappush(heap, (cost + neighbor_cost, neighbor))
    return None
```

```python
def get_route(railway_network, start, end):
    heap = [(0, start, [])]
    visited = set()
    while heap:
        (cost, current, path) = heapq.heappop(heap)
        if current in visited:
            continue
        visited.add(current)
        path = path + [current]
        if current == end:
            return (cost, path)
        for neighbor, neighbor_cost in railway_network[current].items():
            if neighbor not in visited:
                heapq.heappush(heap, (cost + neighbor_cost, neighbor, path))
    return None
```

```python
def main():
    railway_network = load_railway_network()
    start = input("Enter the name of the departure station: ")
    end = input("Enter the name of the destination station: ")
    cost = dijkstra(railway_network, start, end)
    if cost is None:
        print("There is no route between {} and {}.".format(start, end))
        return
    route_cost, route = get_route(railway_network, start, end)
    print("The cheapest cost between {} and {} is {}.".format(start, end, cost))
    print("The route is: {}".format(' -> '.join(route)))
```
